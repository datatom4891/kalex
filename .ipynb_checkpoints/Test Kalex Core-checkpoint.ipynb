```python
from kalex.graph.graph import assemble_kalex
```

```python
kalex = assemble_kalex()
config = {"configurable": {"thread_id": "user_123","user_id":1}}
```

```python
def ask_kalex(user_prompt):
    response = kalex.invoke({"messages":user_prompt}, config=config)
    response_from_kalex = response["messages"][-1]
    return response_from_kalex.content
```

```python
user_prompt = "Who is the current POTUS"
ask_kalex(user_prompt)
```

```python
user_prompt = "That isn't correct, check again"
ask_kalex(user_prompt)
```

```python
user_prompt = "My mistake you were right the first time"
ask_kalex(user_prompt)
```
